```python
from matplotlib import rcParams, cycler
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
dataseturl="https://raw.githubusercontent.com/dataprofessor/data/master/penguins_cleaned.csv"
data=pd.read_csv(dataseturl)
```

```python
data.head(4)
```

```python
data.head()
data1 = data.drop(columns=['species', 'island'])
```

```python
data2 = data[["species"]]
```

```python
data2.head(25)
```

```python
dataspecies = pd.get_dummies(data2)
dataspecies
```

```python
data3 = data[["island"]]
```

```python
data3.head(25)
```

```python
dataisland = pd.get_dummies(data3)
dataisland
```

```python
datanumeric = pd.concat([dataspecies, dataisland], axis=1)
```

```python
Hasil = pd.concat([data1, datanumeric], axis=1)
Hasil
```

min max normalisasi

```python
dataNormalisasi = ['bill_length_mm',	'bill_depth_mm',	'flipper_length_mm',	'body_mass_g']
FrameNormalisasi = Hasil[dataNormalisasi]
FrameNormalisasi
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(FrameNormalisasi)
```

```python
# with open('ScalarMinMax','wb') as r:
#   pickle.dump(scaler,r)
```

```python
hasilNormalisasi = scaler.transform(FrameNormalisasi)
```

```python
hasilNormalisasi = pd.DataFrame(hasilNormalisasi,columns = dataNormalisasi)
hasilNormalisasi
```

```python
Hasil = Hasil.drop(columns = dataNormalisasi)
Hasil = pd.concat([Hasil, hasilNormalisasi], axis=1)
Hasil
```
