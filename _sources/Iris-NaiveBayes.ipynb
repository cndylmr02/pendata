## Implementasi Naive Bayes

```python
from matplotlib import rcParams, cycler
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from collections import OrderedDict
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score, f1_score
from sklearn.datasets import make_classification
import numpy as np
import pandas as pd
```

```python
datairisurl="https://raw.githubusercontent.com/cindylmr19/dataset/main/iris.csv"
iris=pd.read_csv(datairisurl)
```

```python
iris.head(150)
```

```python
data1 = iris.drop(columns=['class'])
data1.head(150)
```

```python
data2 = iris.drop(columns=['sepallength',	'sepalwidth',	'petallength',	'petalwidth'])
data2.head(150)
```

```python
dataNormalisasi = ['sepallength',	'sepalwidth',	'petallength',	'petalwidth']
FrameNormalisasi = data1[dataNormalisasi]
FrameNormalisasi
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(FrameNormalisasi)
```

```python
hasilNormalisasi = scaler.transform(FrameNormalisasi)
```

```python
hasilNormalisasi = pd.DataFrame(hasilNormalisasi, columns = dataNormalisasi)
hasilNormalisasi
```

```python
Hasil = pd.concat([data2, hasilNormalisasi], axis=1)
Hasil
```

```python
percent_amount_of_test_data = 0.3
```

```python
X1 = Hasil.iloc[:,1:5].values
Y1 = Hasil.iloc[:, 0].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X1, Y1, test_size = percent_amount_of_test_data, random_state=0)
```

```python
X_train.shape
```

```python
Y1
```

```python
### Dictionary to store model and its accuracy

model_accuracy = OrderedDict()

### Dictionary to store model and its precision

model_precision = OrderedDict()

### Dictionary to store model and its recall

model_recall = OrderedDict()
```

```python
naive_bayes_classifier = GaussianNB()
naive_bayes_classifier.fit(X_train, y_train)
Y_pred_nb = naive_bayes_classifier.predict(X_test)

cm = confusion_matrix(y_test, Y_pred_nb)

print('Confusion matrix for Gaussian Naive Bayes\n',cm)

naive_bayes_accuracy = round(100 * accuracy_score(y_test, Y_pred_nb), 2)
model_accuracy['Gaussian Naive Bayes'] = naive_bayes_accuracy

naive_bayes_precision = round(100 * precision_score(y_test, Y_pred_nb, average = 'weighted'), 2)
model_precision['Gaussian Naive Bayes'] = naive_bayes_precision

naive_bayes_recall = round(100 * recall_score(y_test, Y_pred_nb, average = 'weighted'), 2)
model_recall['Gaussian Naive Bayes'] = naive_bayes_recall

print('The accuracy of this model is {} %.'.format(naive_bayes_accuracy))
print('The precision of this model is {} %.'.format(naive_bayes_precision))
print('The recall of this model is {} %.'.format(naive_bayes_recall))
```
